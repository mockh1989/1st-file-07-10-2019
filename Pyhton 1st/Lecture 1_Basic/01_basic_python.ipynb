## If, else and elif

```python
0 => MR, 1=> MS, 2=> MRS 3=> Sir/Madam
```

```python
def print_title(code=3):
    if code == 0:
        return  'MR'
    elif code == 1:
        return 'MS'
    elif code == 2:
        return 'MRS'
    else:
        return 'Sir/Madam'
```

```python
print_title(1)
```

# Homework
Write a program to prints out a list of all the divisors of a user input number. 
A divisor is a number that divides into another number without a remainder. 
For example, 1,3,5,15 is the divisors of 15.

```python
# loop from 1 to input number
# for each element (1,2,3....input_number)
# check if it can be divided by input number (i.e. no remainder)
# if no remainder => get element 
# else => skip 
```

```python
def find_divisor(n):
    for x in range(1,n+1):
        if (n%x == 0):
           print(x)
```
