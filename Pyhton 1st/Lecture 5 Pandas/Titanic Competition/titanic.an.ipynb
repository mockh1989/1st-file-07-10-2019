## Feature Engineering

```python
# Fill missing value
# grouping
# string to int
```

```python
# Fill missing value
# Age
# Cabin
# Embarked


mean_age = (df_train['Age'].sum() + df_test['Age'].sum())/(sum(df_train['Age'].notnull())+sum(df_test['Age'].notnull()))
df_train['Age'] = df_train['Age'].fillna(mean_age)
df_test['Age'] = df_test['Age'].fillna(mean_age)
# df_train.loc[df_train['Age'].isnull(),'Age'] = mean_age
```

```python
df_train.info()
```

```python
df_train['Cabin'].unique()
```

```python
df_test['Cabin'].unique()
```

```python
df_train['Cabin'] = df_train['Cabin'].fillna('X')
df_test['Cabin'] = df_test['Cabin'].fillna('X')
```

```python
df_train['Cabin'] = df_train['Cabin'].apply(lambda x:str(x)[0])
df_test['Cabin'] = df_test['Cabin'].apply(lambda x:str(x)[0])
```

```python
# df_train = pd.concat([df_train,pd.get_dummies(df_train['Age'],prefix= 'Age')],axis=1)
df_train = pd.concat([df_train,pd.get_dummies(df_train['Cabin'],prefix='Cabin')],axis=1)
df_train = pd.concat([df_train,pd.get_dummies(df_train['Embarked'],prefix='Embarked')],axis=1)
df_train = pd.concat([df_train,pd.get_dummies(df_train['Sex'],prefix='Sex')],axis=1)
```

```python
# df_train.drop([i for i in df_train.columns if 'Age_' in i],axis=1, inplace=True)
```

```python
label_train = df_train['Survived']
df_train.drop(['PassengerId','Survived','Name','Sex','Ticket','Cabin','Embarked'],axis=1,inplace=True)
```

## Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
X = df_train
y = label_train
```

```python
# create a model
lg = LogisticRegression()
```

```python
# train the model
lg.fit(X,y)
```

```python
# predict
y_predict_proba = lg.predict_proba(X)[:,1]
```

```python
y_predict = lg.predict(X)
```

```python
result = pd.DataFrame([y,y_predict]).T
```

```python
result.columns = ['Actual','Predict']
```
